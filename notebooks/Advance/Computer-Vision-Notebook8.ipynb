```python
import cv2
import matplotlib.pyplot as plt
```

```python
# car = cv2.imread("./multiple cars.jpg")
car = cv2.imread("./2 cars.jpg")
gray_car = cv2.cvtColor(car,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_car,cmap = "gray")
```

```python
plate_cascade = cv2.CascadeClassifier("./haarcascade_russian_plate_number.xml")
```

```python
plates = plate_cascade.detectMultiScale(gray_car, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

```python
for (x, y, w, h) in plates:
    cv2.rectangle(car, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

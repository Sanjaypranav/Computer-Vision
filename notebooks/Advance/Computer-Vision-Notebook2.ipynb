```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_1 = cv2.imread("C:\Kishore\Studies\Sem 9\CV Lab\Joy Balde Runner 2049.png")
image_2 = cv2.imread("C:\Kishore\Studies\Sem 9\CV Lab\Joy Balde Runner 2049 - Rotated.png")

image_1_gray = cv2.cvtColor(image_1,cv2.COLOR_BGR2GRAY)
image_2_gray = cv2.cvtColor(image_2,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_1_gray,cmap = "gray")
```

```python
plt.imshow(image_2_gray,cmap = "gray")
```

```python
# ORB Detector

detect = cv2.ORB_create()
key_point1,descrip1 = detect.detectAndCompute(image_1_gray,None)
key_point2,descrip2 = detect.detectAndCompute(image_2_gray,None)
```

```python
# BF
brute_force = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
no_of_matches = brute_force.match(descrip1,descrip2)

no_of_matches = sorted(no_of_matches,key = lambda x:x.distance)
```

```python
tot_feature_matches = len(no_of_matches)
tot_feature_matches
```

```python
output_image = cv2.drawMatches(image_1_gray,key_point1,image_2_gray,key_point2,no_of_matches[:30],None,flags = 2)
```

```python
plt.imshow(output_image)
```
